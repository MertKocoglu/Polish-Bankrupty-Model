```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE

# Models
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier, HistGradientBoostingClassifier, BaggingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Fill NA
data.fillna(data.mean(), inplace=True)
X = data.drop(['class', 'year'], axis=1)
y = data['class']

# Scoring metrics
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score),
}

# Models
models = {
    'RandomForestClassifier': RandomForestClassifier(random_state=42),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=42),
    'AdaBoostClassifier': AdaBoostClassifier(random_state=42),
    'ExtraTreesClassifier': ExtraTreesClassifier(random_state=42),
    'HistGradientBoostingClassifier': HistGradientBoostingClassifier(random_state=42),
    'BaggingClassifier': BaggingClassifier(random_state=42),
    'XGBClassifier': XGBClassifier(random_state=42),
    'LGBMClassifier': LGBMClassifier(random_state=42, verbose=-1),
    'CatBoostClassifier': CatBoostClassifier(random_state=42, verbose=False),
    'LogisticRegression': LogisticRegression(random_state=42, max_iter=1000),
    'SVC': SVC(random_state=42),
}

# Cross-validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}

for name, model in models.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])
    
    cv_results = cross_validate(pipeline, X, y, cv=kfold, scoring=scoring, n_jobs=-1)
    
    results[name] = {metric: scores.mean() for metric, scores in cv_results.items() if 'test_' in metric}
    print(f"{name}:")
    for metric, score in results[name].items():
        print(f"  {metric}: {score:.4f}")
    print("-----------------------")

# En iyi model
best_model = max(results.items(), key=lambda item: item[1]['test_f1'])
print(f"Best model based on F1 score: {best_model[0]}")
print("Scores:", best_model[1])
```

```python
import optuna
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import StratifiedKFold
import matplotlib.pyplot as plt

# Bağımlı ve bağımsız değişkenler
X = data.drop(['class', 'year'], axis=1)
y = data['class']

# Skorları 5-fold cross-validation ile ölçelim
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def objective(trial):
    params = {
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'depth': trial.suggest_int('depth', 4, 10),
        'l2_leaf_reg': trial.suggest_int('l2_leaf_reg', 1, 10),
        'iterations': trial.suggest_int('iterations', 100, 500),
        'random_state': 42,
        'verbose': False
    }

    # Pipeline: Scaling → SMOTE → CatBoost
    pipeline = ImbPipeline([
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)),
        ('classifier', CatBoostClassifier(**params))
    ])

    # F1 skoruna göre cross-validation
    score = cross_val_score(pipeline, X, y, cv=cv, scoring='f1').mean()
    return score

# Optuna ile en iyisini ara
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=40, timeout=7200)  # 2 saat = 7200 saniye

# Sonuçları yazdır
print("En iyi parametreler:", study.best_params)
print("En iyi F1 skoru:", study.best_value)

# En iyi sonuçları plotla
trials = study.trials_dataframe()
plt.figure(figsize=(10,6))
plt.plot(trials['number'], trials['value'], label='F1 Score')
plt.xlabel('Trial')
plt.ylabel('F1 Score')
plt.title('Optuna Optimization - F1 Score Across Trials')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_predict

# En iyi parametreleri kullanarak pipeline'ı kur
best_params = study.best_params

pipeline = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('classifier', CatBoostClassifier(**best_params, random_state=42, verbose=False))
])

# Cross-validated tahminleri al
y_pred = cross_val_predict(pipeline, X, y, cv=cv)

# Performans ölçümlerini yazdır
print("\nClassification Report:")
print(classification_report(y, y_pred, digits=4))

# Confusion matrix'i çiz
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from lime.lime_tabular import LimeTabularExplainer
import matplotlib.pyplot as plt

# 1. VERİYİ YÜKLE
data = pd.read_csv("data.csv")  # Dosya adını kendi dosyana göre değiştir
X = data.drop(columns=["class"])  # 'class' target sütunu olmalı
# Eksik değerleri doldur
X = X.fillna(X.mean())

y = data["class"]

# 2. EĞİTİM VE TEST AYRIMI
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 3. NORMALİZASYON
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. CATBOOST MODELİ
model = CatBoostClassifier(verbose=0)
model.fit(X_train_scaled, y_train)

# 5. LIME EXPLAINER
explainer = LimeTabularExplainer(
    training_data=X_train_scaled,
    feature_names=X.columns.tolist(),
    class_names=['Non-Bankrupt', 'Bankrupt'],
    mode='classification',
    discretize_continuous=False  # BU SATIRI EKLE
)


# 6. RASTGELE BİR TEST VERİSİ SEÇ
i = 10  # 0- X_test'in uzunluğu arasında rastgele bir sayı
exp = explainer.explain_instance(
    X_test_scaled[i],
    model.predict_proba,
    num_features=10
)

# 7. GÖSTER
# Açıklamayı HTML dosyası olarak kaydet
exp.save_to_file("lime_explanation.html")

# veya .save_to_file("lime_explanation.html") şeklinde kaydedebilirsin
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from lime.lime_tabular import LimeTabularExplainer
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("C:/Users/Mert Koçoğlu/Desktop/polishbank/data.csv")

# Feature-target separation
X = data.drop("class", axis=1)
y = data["class"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Load your CatBoost model (adjust path if needed)
model = joblib.load("catboost_best_model.pkl")  # or full path if needed

# Feature and class names
feature_names = X.columns.tolist()
class_names = ["No", "Yes"] if y.nunique() == 2 else list(y.unique())
```

```python
explainer = LimeTabularExplainer(
    training_data=X_train.values,
    feature_names=feature_names,
    class_names=class_names,
    mode='classification',
    discretize_continuous=False  # HATAYI ENGELLER
)

# Instance explanation
idx = 0
instance = X_test.values[idx]
exp = explainer.explain_instance(instance, model.predict_proba, num_features=10)

# Çıktı
print("Prediction probabilities:", model.predict_proba([instance]))
print("True class:", y_test.iloc[idx])
print("Explanation:")
for feature, weight in exp.as_list():
    print(f"{feature}: {weight:.4f}")
```

```python
# NaN içermeyen örnekleri seç
clean_X_test = X_test.dropna()
clean_y_test = y_test.loc[clean_X_test.index]

# Sınıflara göre indeks seç
yes_idx = clean_y_test[clean_y_test == 1].index[0]
no_idx = clean_y_test[clean_y_test == 0].index[0]

# LIME açıklamaları
exp_yes = explainer.explain_instance(clean_X_test.loc[yes_idx].values, model.predict_proba, num_features=10)
exp_no = explainer.explain_instance(clean_X_test.loc[no_idx].values, model.predict_proba, num_features=10)

# Yazdır
print("== Class: Yes ==")
for feature, weight in exp_yes.as_list():
    print(f"{feature}: {weight:.4f}")

print("\n== Class: No ==")
for feature, weight in exp_no.as_list():
    print(f"{feature}: {weight:.4f}")
```

```python
from sklearn.inspection import permutation_importance

# Run permutation importance on test set
result = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=42)

# Sort and plot
import matplotlib.pyplot as plt

sorted_idx = result.importances_mean.argsort()[::-1][:10]
plt.figure(figsize=(10, 5))
plt.barh(range(len(sorted_idx)), result.importances_mean[sorted_idx], align="center")
plt.yticks(range(len(sorted_idx)), [X_test.columns[i] for i in sorted_idx])
plt.xlabel("Mean Importance (Drop in F1 or Accuracy)")
plt.title("Permutation Feature Importance")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
